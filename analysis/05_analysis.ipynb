```python
import pandas as pd
import os
import math
import plotly.express as px
import plotly.graph_objects as go
```

```python
users = pd.read_csv("./results/users.csv")

folder_path = './data/tracking_files/'
list_of_filenames = os.listdir(folder_path)

tasks = ["A1", "A2", "A3", "A4", "A5", "A6", "A7", 
         "B1", "B2", "B3", "B4", 
         "C1", "C2", "C3", "C4", 
         "D1", "D2", 
         "E1", "E2", "E3", 
         "F1", "F2", "F3"]
         
users
```

# path

```python
for task in tasks:
    
    task_dict = {}

    for file in list_of_filenames:
        if file.split("_")[1] == task + ".csv":
            df = pd.read_csv("data/tracking_files/" + file)
            task_dict[file.split("_")[0]] = df

    path_lengths = []

    for user_nr in users["nr"]:
        
        path_length = 0
        df_path = task_dict[str(user_nr)]

        for i in range(1, len(df_path)):
            dx = df_path['x'].iloc[i] - df_path['x'].iloc[i-1]
            dy = df_path['y'].iloc[i] - df_path['y'].iloc[i-1]
            segment_length = math.sqrt(dx**2 + dy**2)
            path_length += segment_length
        path_lengths.append(path_length)

    users["path_" + task] = path_lengths

users
```

# time

```python
for task in tasks:
    
    task_times = []

    for user in users["nr"]:
        df = pd.read_csv("data/tracking_files/" + str(user) + "_" + task + ".csv")
        task_times.append(df.iloc[-1]["ms"] / 1000)

    users["time_" + task] = task_times

users
```

# DuGe

```python
for task in tasks:
    users["v_" + task] = users["path_" + task] / users["time_" + task] 

users
```

# Anzahl Clicks

```python
for task in tasks:
    
    task_num_clicks = []

    for user in users["nr"]:
        df = pd.read_csv("data/tracking_files/" + str(user) + "_" + task + ".csv")
        task_num_clicks.append(df['action'].value_counts()['c'])

    users["num_clicks_" + task] = task_num_clicks

users
```

```python
columns = [elem for elem in users.columns.tolist() if elem.startswith("num_clicks_")]
users[columns].value_counts()
```

# Zeit in Ruhe

```python
for task in tasks:
    
    task_dict = {}

    for file in list_of_filenames:
        if file.split("_")[1] == task + ".csv":
            df = pd.read_csv("data/tracking_files/" + file)
            task_dict[file.split("_")[0]] = df

    times_ruhe = []

    for user_nr in users["nr"]:
        
        time_ruhe = 0
        df_path = task_dict[str(user_nr)]

        for i in range(1, len(df_path)):
            dms = df_path['ms'].iloc[i] - df_path['ms'].iloc[i-1]
            if dms > 500:
                time_ruhe += dms
        times_ruhe.append(time_ruhe / 1000)

    users["time_ruhe_" + task] = times_ruhe
    users["time_ruhe_pr_" + task] = users["time_ruhe_" + task] / users["time_" + task] * 100

users
```

# Aufgaben mit Fertig-Button

```python
tasks = ["A1", "A2", "A3", "A4", "A5", "A6", "A7",
         "C1", "C2", "C3", "C4", 
         "F1", "F2", "F3"]
```

```python
for task in tasks:
    
    task_dict = {}

    for file in list_of_filenames:
        if file.split("_")[1] == task + ".csv":
            df = pd.read_csv("data/tracking_files/" + file)
            task_dict[file.split("_")[0]] = df

    path_lengths = []
    card_click_times = []

    for user_nr in users["nr"]:

        df_path = task_dict[str(user_nr)]
        card_click_time = df_path[df_path["action"] == "c"].iloc[-2]["ms"]
        card_click_times.append(card_click_time)

    print(card_click_times)    

    for user_nr in users["nr"]:
        
        path_length = 0
        df_path = task_dict[str(user_nr)]

        for i in range(1, len(df_path)):
            if df_path["ms"].iloc[i] < card_click_times[user_nr-1]:
                dx = df_path['x'].iloc[i] - df_path['x'].iloc[i-1]
                dy = df_path['y'].iloc[i] - df_path['y'].iloc[i-1]
                segment_length = math.sqrt(dx**2 + dy**2)
                path_length += segment_length
        path_lengths.append(path_length)

    if task.startswith("C"):
        users["path_to_click_" + task] = path_lengths
        
    users["time_to_click_" + task] = card_click_times
    users["time_to_click_" + task] = users["time_to_click_" + task] / 1000
    users["time_after_click_" + task] = users["time_" + task] - users["time_to_click_" + task]

users
```

```python
columns = [elem for elem in users.columns.tolist() if elem.startswith("path_to_click_")]
users[columns].describe()
```

```python
columns = [elem for elem in users.columns.tolist() if elem.startswith("time_to_click_")]
users[columns].describe()
```

```python
columns = [elem for elem in users.columns.tolist() if elem.startswith("time_after_click_")]
users[columns].describe()
```

# Abweichung Idealpfad für Aufgabengruppe C

```python
start = (715,75)

# Koordinaten (x,y) von C1, C2, C3, C4
c_coords = [(135,485),(1125,640),(350,445),(800,635)]

idealpfade = []

for task_coord in c_coords:
    dx = task_coord[0] - start[0]
    dy = task_coord[1] - start[1]
    segment = math.sqrt(dx**2 + dy**2)
    print(dx, dy, segment)
    idealpfade.append(segment)

idealpfade
```

```python
for i in range(4):
    users["path_to_click_-Idealpfad_C" + str(i+1)] = (users["path_to_click_C" + str(i+1)] / idealpfade[i] - 1) * 100

users
```

```python
columns = [elem for elem in users.columns.tolist() if elem.startswith("path_to_click_-Idealpfad_C")]
users[columns].describe()
```
